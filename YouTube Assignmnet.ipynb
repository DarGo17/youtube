```python
cleaned_df = us_yt_df[["trending_date", "category_id", "publish_time", "views", "likes"]]
sorted_df = cleaned_df.sort_values(by="category_id", ascending=True)
sorted_df
```

```python
yt_json = pd.read_json("US_category_id.json")
```

```python
yt_json
```

```python

import pandas as pd

code_df = pd.DataFrame(yt_json["items"].apply(lambda x: {
    "id": x["id"],
    "title": x["snippet"]["title"]
}).tolist())

print(code_df)

```

```python
# code_df["title"].isin(cleaned_df["category_id"])
```

```python
sorted_df["title"] = code_df["title"]
sorted_df
```

```python
code_df
```

```python
# Make sure the id column in code_df is the same type as category_id in sorted_df
code_df['id'] = code_df['id'].astype(str)
sorted_df['category_id'] = sorted_df['category_id'].astype(str)

# Create a mapping dictionary from code_df
code_map = code_df.set_index('id')['title']

# Map the numeric codes in sorted_df to their titles
sorted_df['category_title'] = sorted_df['category_id'].map(code_map)
```

```python
sorted_df
```

```python
smaller_df = sorted_df.sample(1000)
smaller_df
```

```python
smaller_df.drop(["title"], axis =1)
```

```python
fig, ax = plt.subplots(figsize=(25,6))
ax.hist(smaller_df["category_title"], bins=15)
ax.set_xticklabels(ax.get_xticklabels(), rotation = 45)
plt.tight_layout()


ax.set_xlabel("Entertainment Category")
ax.set_ylabel("Views")
ax.set_title("2017 - 2018 YouTuber Entertainment")

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(25,6))
ax.bar(smaller_df["category_title"], smaller_df["views"])
plt.tight_layout()
ax.set_xlabel("Entertainment Category")
ax.set_ylabel("Number of Views")
ax.set_title("YouTube Entertainment")
plt.tight_layout()
plt.show()
```
